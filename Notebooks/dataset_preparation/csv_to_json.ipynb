```python
import pandas as pd
import numpy as np
import json
```

```python
df1=pd.read_csv("./cleaned_data_2 (1).csv")
df2=pd.read_csv("./leetcode.csv")
df3=pd.read_csv("./taco_cleaned.csv")
```

```python
df1=df1[['instruction','output']]
```

```python
df1.head()
```

```python
df2.head()
```

```python
df2=df2[["description","code"]]
```

```python
df2.head()
```

```python
df3=df3[['question','solutions']]
```

```python
df3.head()
```

```python
df1=df1.values
df2=df2.values[200:]
df3=df3.values
```

```python
df1.shape
```

```python
df2.shape
```

```python
df3.shape
```

```python
df4=np.vstack((df1,df2))
df=np.vstack((df4,df3))
```

```python
df.shape
```

```python
json_f={}
index=1
for i in df:
  x={
      "instruction":i[0],
      "code":i[1]
  }
  json_f[f"question_{index}"]=x

  index+=1


```

```python
json_f
```

```python
with open("csv_to_json.json","w") as f:
  json.dump(json_f,f,indent=4)
```

```python
with open("csv_to_json.json","r") as f:
  data=json.load(f)

with open("csv_to_jsonl.jsonl","w") as f:
  for key,value in data.items():
    f.write(json.dumps(value) + "\n")
```
