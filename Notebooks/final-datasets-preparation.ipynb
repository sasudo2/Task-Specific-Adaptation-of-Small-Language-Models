```python
df1 = pd.read_csv("/kaggle/input/taco-cleaned/taco_cleaned.csv")
df2 = pd.read_csv("/kaggle/input/cleaned-datasets-of-alpaca/cleaned_data_2.csv")
df3 = pd.read_csv("/kaggle/input/leetcode-with-desciption-and-code/leetcode_with_description_and_code.csv")
```

```python
df3 = df3.iloc[200:1201]
```

```python
df3 = df3[['description', 'code']]
```

```python
df2 = df2.iloc[:5000]
```

```python
df2.head()
```

```python
df2 = df2[['instruction','output']]
```

```python
df1 = df1.iloc[:19000]
```

```python
df1.head()
```

```python
df1 = df1[['question','solutions']]
```

```python
df1 = df1.rename(columns={'question': 'instruction', 'solutions': 'code'})
df2 = df2.rename(columns={'instruction': 'instruction', 'output': 'code'})
df3 = df3.rename(columns={'description': 'instruction', 'code': 'code'})
```

```python
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
combined_df.head()
```

```python
combined_df.info()
```

```python
shuffled_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
train = shuffled_df.iloc[0:18000]
```

```python
validation = shuffled_df.iloc[18000:21500]
```

```python
test = shuffled_df.iloc[21500:]
```

```python
test.head()
```

```python
import json  # <- this was missing

# Write to JSONL
with open("train.jsonl", "w", encoding="utf-8") as f:
    for _, row in train.iterrows():
        obj = {
            "instruction": str(row['instruction']),
            "code": str(row['code'])
        }
        f.write(json.dumps(obj, ensure_ascii=False) + "\n")

print("JSONL file created: train.jsonl")
```
