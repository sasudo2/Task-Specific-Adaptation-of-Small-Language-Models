```python
import pandas as pd

df = pd.read_csv("/content/train.csv")

df.head()
```

```python
import ast

def extract_single_solution(sol):
    if pd.isna(sol):
        return None


    try:
        parsed = ast.literal_eval(sol)
        if isinstance(parsed, list) and len(parsed) > 0:
            return str(parsed[0])
    except:
        pass


    for sep in [",", "||", "\n", ";"]:
        if sep in sol:
            return sol.split(sep)[0].strip()


    return sol.strip()
```

```python
df_clean = pd.DataFrame({
    "question": df["question"],
    "solution": df["solutions"].apply(extract_single_solution)
})
```

```python
df_clean.head()
```

```python
output_path = "/content/clean_dataset_single_solution.csv"
df_clean.to_csv(output_path, index=False)

output_path
```
