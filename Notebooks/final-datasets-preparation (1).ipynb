````python
def complexity_filter(df, text_col='code', min_words=7, max_repeat_ratio=0.6):
    df = df[df[text_col].str.split().str.len() > min_words]
    def repetition_ratio(text):
        words = text.split()
        return 1.0 - (len(set(words)) / len(words)) if len(words) > 0 else 0
    df = df[df[text_col].apply(repetition_ratio) < max_repeat_ratio]
    return df

combined_df = complexity_filter(combined_df)
````

````python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import MiniBatchKMeans
import numpy as np

def diversity_sampling(df, text_col='instruction', num_clusters=21000):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = model.encode(
        df[text_col].tolist(),
        show_progress_bar=True,
        normalize_embeddings=True
    )

    kmeans = MiniBatchKMeans(
        n_clusters=num_clusters,
        batch_size=1024,
        random_state=42
    )
    
    clusters = kmeans.fit_predict(embeddings)
    df['cluster'] = clusters

    # Compute distance to centroid
    distances = np.linalg.norm(
        embeddings - kmeans.cluster_centers_[clusters],
        axis=1
    )

    df['distance'] = distances
    print(df['distance'].describe())
    curated_df = df.loc[df.groupby('cluster')['distance'].idxmin()]

    return curated_df.drop(columns=['cluster', 'distance'])
combined_df = diversity_sampling(combined_df)
````
