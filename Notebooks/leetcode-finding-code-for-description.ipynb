```python
import os
import re
import pandas as pd

CODE_DIR = "/kaggle/input/leetcode/leetcode-dataset/data"


def normalize(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9]+", "-", text)
    return text.strip("-")


code_map = {}

for filename in os.listdir(CODE_DIR):
    if filename.endswith(".py"):
        file_path = os.path.join(CODE_DIR, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            code_map[normalize(filename[:-3])] = f.read()


df["code"] = df["title"].apply(
    lambda title: code_map.get(normalize(title))
)


print("Total rows:", len(df))
print("Code matched:", df["code"].notna().sum())


df.to_parquet("leetcode_with_code.parquet", index=False)
df.to_csv("leetcode_with_code.csv", index=False)
```
