````python
import pandas as pd
import numpy as np
````

````python
df1=pd.read_csv("13001_22000.csv")
df2=pd.read_csv("5000_13000.csv")
````

````python
df1.head()
````

````python
df1=df1[["instruction",'output','category']].values
df2=df2[['instruction','output','category']].values
````

````python
df=np.concatenate((df1,df2))
````

````python
df.shape
````

````python
df=pd.DataFrame(df,columns=['instruction','output','category'])
````

````python
df.head()
````

````python
valid_categories=df['category'].value_counts()

valid_categories=valid_categories[valid_categories>5].index
````

````python
df=df[df['category'].isin(valid_categories)]
````

````python
df.head()
````

````python
allowed_categories=["array", "string", "math", "sorting_searching", "file", "heap","geometry", "hashing", "two_pointers", "dynamic_programming", "regex", "linked_list", "recursion", "tree", "stack", "graph", "bit_manipulation", "threading"]
````

````python

array_miss_matched=['array, string','string, array','array, two_pointers','matrix_grid',]
string_miss_matched=['string, two_pointers','json','string, two_pointers, sliding_window','string, bit_manipulation']
math_miss_matched=['math, number_theory','math, data_analysis','math, array',"array, math"]
hashing_miss_matched=['dictionary','array, hashing',"string, hashing"]
tree_miss_matched=['parsing','binary_search']
set_miss_matched=['set_operations','set, array']
dp_miss_matched=['array, dynamic_programming','permutations','math, dynamic_programming']
searching_miss_matched=["searching",'linear_search','binary_search','array, sorting_searching','string, sorting_searching','sorting\\_searching',]


c=df.values
for i in range(len(c)):
  if c[i][2] in array_miss_matched:
    c[i][2]='array'

  if c[i][2] in string_miss_matched:
    c[i][2]='string'

  if c[i][2] in math_miss_matched:
    c[i][2]='math'

  if c[i][2] in hashing_miss_matched:
    c[i][2]='hashing'

  if c[i][2] in tree_miss_matched:
    c[i][2]='tree'

  if c[i][2] in set_miss_matched:
    c[i][2]='set'

  if c[i][2] in dp_miss_matched:
    c[i][2]='dynamic_programming'

  if c[i][2] in searching_miss_matched:
    c[i][2]='sorting_searching'
````

````python
df=pd.DataFrame(c,columns=['instruction','code','category'])
````

````python
df['category'].value_counts()
````

````python
for i,row in df.iterrows():
  if row[2] not in allowed_categories:
    df.drop(i, axis=0,inplace=True)
````
